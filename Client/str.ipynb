```python
def format_string(s):
    # Determine the length of the input string
    n = len(s)

    # Calculate the number of asterisks to put on each side of the input string
    # to center it on the second line
    asterisks = max(0, (34 - n) // 2)

    # Construct the formatted string
    num_asterisks = 2 * (asterisks + 2) + n
    top_bottom = "*" * num_asterisks 
    middle = "*" * (asterisks + 2) + s + "*" * (asterisks + 2 + n % 2)
    middle = middle.center(num_asterisks) + "\n"
    result = top_bottom + "\n" +  middle + top_bottom
    return "/"+result+"/"
```

```python
s = "FUNCTION POUR GERER LE TEMPS D'EXECUTION DU CLIENT".upper()
print(format_string(s))
```
